```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import lcapy
```

```python
C1 = 0.25e-9
C2 = 20e-9
C3 = 20e-9
R1 = 250e3
R2 = 1e6
R3 = 25e3
R4 = 56e3
```

```python

cct = lcapy.Circuit(f"""
C1 1_2 2 {C1}; right
R1a 2 3; down
R1b 3 4; down
R2 4 5; down
R3a 5 6; down
R3b 6 0_2; down
R4 1_2 7 {R4}; down
C2 7 4 {C2}; right
W 7 7_2; down
C3 7_2 6 {C3}; right
W 0 0_2; right
W 1 1_2; right
W 3 3_1; right
W 0_2 0_3; right
P1 3_1 0_3; down
""")

cct.draw()
```

```python
H = cct.transfer(1, 0, 3, 0)
```
